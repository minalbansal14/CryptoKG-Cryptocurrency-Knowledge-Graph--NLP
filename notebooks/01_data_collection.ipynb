# Activity 1 — Data Collection

Curate a domain-specific corpus of English sentences covering relationships
between cryptocurrency and fintech entities.

**Goal:** Produce `data/sentences.txt`, `data/entities_dict.json`, and `data/relations_dict.json`.

## 1. Setup

```python
# Install dependencies (Colab)
# !pip install spacy
# !python -m spacy download en_core_web_lg

import json, pathlib

DATA_DIR = pathlib.Path('../data')
DATA_DIR.mkdir(exist_ok=True)
```

## 2. Domain Sentences

Paste your curated sentences below (or load from `data/sentences.txt`).

```python
# TODO: replace with your full corpus from the Colab notebook
sentences = [
    "Ripple partnered with Visa to improve their cross-border payment system.",
    "Ethereum was developed by Vitalik Buterin.",
    # ... add more sentences
]

print(f"Loaded {len(sentences)} sentences")
```

## 3. Entity Gazetteer

Map each known entity name to its Wikidata Q-ID.

```python
# TODO: populate from your Colab notebook
entities_dict = {
    "Ripple":          "Q1307473",
    "Visa":            "Q328840",
    "Ethereum":        "Q11378511",
    "Vitalik Buterin": "Q16082920",
    # ...
}

with open(DATA_DIR / 'entities_dict.json', 'w') as f:
    json.dump(entities_dict, f, indent=2)
print(f"Saved {len(entities_dict)} entities")
```

## 4. Relation Gazetteer

Map each known relation phrase to its Wikidata P-ID.

```python
# TODO: populate from your Colab notebook
relations_dict = {
    "partnered with": "P1327",
    "invested in":    "P1344",
    "developed by":   "P61",
    # ...
}

with open(DATA_DIR / 'relations_dict.json', 'w') as f:
    json.dump(relations_dict, f, indent=2)
print(f"Saved {len(relations_dict)} relations")
```

## 5. Save Sentences

```python
with open(DATA_DIR / 'sentences.txt', 'w') as f:
    f.write('\n'.join(sentences))
print(f"Saved {len(sentences)} sentences to data/sentences.txt")
```
